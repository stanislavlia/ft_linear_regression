```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import json

df = pd.read_csv("../data.csv")
```

### Distribution of Price and Distance

```python

df["price"].plot.hist(
    bins=10,          # number of bins; tweak as needed
    alpha=0.7,        # transparency
    figsize=(8, 5),   # figure size in inches
    grid=True,
    density=True,
    color="orange"         # show grid lines behind bars

)

plt.xlabel('Value of price')
plt.ylabel('Density')
plt.title('Histogram of Price')
plt.show()
```

```python

df["km"].plot.hist(
    bins=10,          # number of bins; tweak as needed
    alpha=0.7,        # transparency
    figsize=(8, 5),   # figure size in inches
    grid=True,
    density=True         # show grid lines behind bars

)

plt.xlabel('Value of Distance(KM)')
plt.ylabel('Density')
plt.title('Histogram of Distance')
plt.show()
```

### Distance vs Price Plot (Bonus)


```python
plt.figure(figsize=(10, 6))
plt.scatter(df["km"], df["price"], color="blue")
plt.ylim(0, df["price"].max() * 1.2)
plt.xlabel("Distance in KM", fontsize=12)
plt.ylabel("Price of Car", fontsize=12)
plt.title("Mileage vs Price Plot", fontsize=14)
```

### Plot for Linear Model (Bonus)

```python
###Load Parameters

param_path = "../linreg_params.json"

with open(param_path, "r") as f:
    params = json.load(f)

print("Learned Pamateres:")
params
```

```python
y_predicts = params["bias"] + params["theta1"] * df["km"]


plt.figure(figsize=(12, 8))
plt.scatter(df["km"], df["price"], color="blue")
plt.plot(df["km"], y_predicts, color="red",  label="linear regression model")
plt.ylim(0, df["price"].max() * 1.2)
plt.text(x=150_000, y=8000, s=f"Linear Model: Mileage = {params['bias']:.2f} + {params['theta1']:.2f} * Mileage")
plt.xlabel("Distance in KM", fontsize=12)
plt.ylabel("Price of Car", fontsize=12)
plt.title("Linear Model For Car Price", fontsize=14)
```
